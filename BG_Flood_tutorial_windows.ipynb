# **BG_Flood tutorial workshop**
BG-Flood is an open-source hydrodynamics solver used to simulate rainfall, fluvial, coastal and tsunami triggered inundations.
The source code and released versions of the solver are available on the dedicated GitHub repository.
> BG-Flood GitHub repository: https://github.com/CyprienBosserelle/BG_Flood/

The installation procedure varies depending on whether the working platform is MS-Windows, Mac, or Unix/Linux.
> If needed, see Manual for installation instructions: https://cyprienbosserelle.github.io/BG_Flood/Install/

---

### **1. BG-Flood for MS-Windows**
This workshop will be run using the windows version of BG-Flood. A pre-compiled version is directly accessible on the GitHub releases.

Download and extract the release v1.0.
- Go to https://github.com/CyprienBosserelle/BG_Flood/releases/
- Download "BG_Flood-v1.0.zip"
- Extract the file
- Double click the executable, it should open and close a terminal page
> If the "Windows protected your PC" page opens, click on "More info" and "Run anyway"
---

### **2. Post-processing and visualisation toolkit creation**
BG-Flood results can be visualised using multiple toolkits.
- GIS visualisation softwares: **QGIS**, ArcGIS
- Scientific coding languages: Python, Julia, R
- Geospatial data processing libraries/packages: GDAL, NCO, GMT, CDO

This workshop will focus on QGIS and basic Python post-processing utilities.

> Visualisation and post-processing tools are **not** included in the BG-Flood repository at this stage but will be included in the near future. Contact joe.pelmard@auckland.ac.nz or alice.harang@earthsciences.nz if needed.

### **3. Python post-processing setup**
For post-processing we will be focusing on a standard python data-processing packages. To do so, we are going to create a virtual environment (venv) where the packages will be stored using the **conda** package and environment manager.
> For details about conda and venvs, see https://docs.conda.io/projects/conda/en/latest/user-guide/getting-started.html

- To create a new conda environment, Use the following commands in the terminal
```bash
conda init
conda create -n BG-Flood-post python=3.10 -y
conda activate BG-Flood-post
```

- Now install the required packages
```bash
conda install -c conda-forge -y numpy xarray rioxarray matplotlib contextily
```
> *contextily* is a package to retrieve tile maps from the internet. It can add those tiles as basemap to matplotlib figures (e.g. OpenStreetMaps)
